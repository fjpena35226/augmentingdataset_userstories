```
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np
import csv
from datetime import datetime
import codecs
import random 
```

```
env_url = '/content/drive/My Drive/TP-final-SI'
file_url = env_url + '/dataSetUserStoriesRecognition.csv'

######LOADING FILE
csvfile = codecs.open(file_url, 'r', encoding='utf-8', errors='ignore')
reader = csv.DictReader(csvfile, delimiter=';')

data_user_stories = []

for row in reader:
  if int(row['Class'].lower()) == 1:
    data_user_stories.append(row['Resumen'].lower())  

csvfile.close()
```

```
max_lenght = 25
trunc_type = 'post'

tokenizer = Tokenizer()

#####TOKENIZING VOCAB
tokenizer.fit_on_texts(data_user_stories)
total_words = len(tokenizer.word_index) + 1
word_index = tokenizer.word_index

#print(word_index)

####TOKENIZING TRAINING SEQUENECES
sequences = tokenizer.texts_to_sequences(data_user_stories)
padded = pad_sequences(sequences, maxlen=max_lenght, truncating=trunc_type)

reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

def decode_seq(text):
  return ' '.join([reverse_word_index.get(i,'') for i in text])
```

```
seq = padded[0] 
print("Sequence: " + format(seq))
print("String: " + format(decode_seq(seq)).strip())
```

```
new_negative_examples = []
for i in range(1000):
  r_index = random.randint(0,len(padded) - 1)
  seq_padded = padded[r_index]
  string2 = ""
  c = 0
  while string2 == "" and c < 10:
    r_split = random.randint(1, int(max_lenght * 0.45))
    splited_seq1 = seq_padded[r_split:]
    splited_seq2 = seq_padded[:r_split]
    string1 = format(decode_seq(splited_seq1)).strip()
    string2 = format(decode_seq(splited_seq2)).strip()
    c = c + 1
  new_negative_examples.append(string1)
  new_negative_examples.append(string2)

for ex in new_negative_examples:
  print(ex)
```

```
csvfile = open( env_url + '/new_tunned_examples.csv',"w+")

fieldnames = ['Resumen','Class']
writer = csv.DictWriter(csvfile, fieldnames=fieldnames,delimiter=';')
writer.writeheader()
for resumen in new_negative_examples:   
  if resumen != "":
    writer.writerow({'Resumen': resumen, 'Class': 0})
    
csvfile.close()
print("Results saved successfuly")    
```
